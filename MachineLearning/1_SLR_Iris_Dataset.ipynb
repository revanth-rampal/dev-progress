# Simple Linear Regression Using Iris Dataset


```python
import seaborn as sns
```

#### Loading the dataset

```python
iris = sns.load_dataset('iris')
iris
```

```python
iris =iris[['petal_length','petal_width']]
```

```python
iris
```

```python
X= iris['petal_length']
y= iris['petal_width']
```

#### Data Visualisation of the data

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,y)
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.show()
```

#### Splitting the dataset for Training and Testing

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=23)
```

```python
X_train
```

#### reshaping to 2 dimentional data|

```python
import numpy as np
X_train = np.array(X_train).reshape(-1,1)
X_test = np.array(X_test).reshape(-1,1)
```

### importing linear regression model

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

### Training the data for linear regression

```python
lr.fit(X_train,y_train)
```

```python
c=lr.intercept_
```

```python
m= lr.coef_
```

```python
lr.predict([[5.1]])
```

```python
plt.scatter(X,y)
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.plot(X,m*X+c ,color='red')
plt.show()
```

#### Finding the R2 score


1: Perfect fit (all data points lie on the regression line).

```python

from sklearn.metrics import r2_score

y_pred = lr.predict(X_test)
r2 = r2_score(y_test, y_pred)
print("R-squared value:", r2)
```

### Finding the MSE (mean Squared Error)

 Lower values are better, with MSE=0MSE=0 indicating a perfect prediction.

```python
from sklearn.metrics import mean_squared_error
mse= mean_squared_error(y_test,y_pred)
mse
```
