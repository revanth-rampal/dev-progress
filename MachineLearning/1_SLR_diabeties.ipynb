```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_diabetes
diabeties = load_diabetes()
diabeties
```

```python
df=pd.DataFrame(diabeties.data,columns=diabeties.feature_names)
df
```

```python
df = df[['bmi','bp']]
df.head()
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
plt.scatter(X,y)
plt.xlabel('bmi')
plt.ylabel('bp')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train,y_train)
```

```python
c=reg.coef_
```

```python
m=reg.intercept_
```

```python
plt.scatter(X,y)
plt.xlabel('bmi')
plt.ylabel('bp')
plt.plot(X,c*X+m,color='red')
```

```python
from sklearn.metrics import r2_score
y_pred = reg.predict(X_test)
score = r2_score(y_test,y_pred)
score
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test,y_pred)
mse
```

**The model has a very low R2R2 score (0.063), indicating it explains only 6.3% of the variance in the target variable. The MSE (0.0021) suggests the prediction errors are small in magnitude, but the model's performance is poor in terms of explaining data variability.**
